```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure , output_file, show
from bokeh.sampledata.iris import flowers
output_file('test.html')

p = figure(title = 'test flower')
p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"
p.circle(flowers['petal_length'] , flowers['petal_width'])
show(p)
```

```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
```

```python
#In Bokeh, glyphs are the basic visual building blocks for creating plots. Glyphs are visual shapes or markers that can be used to represent data points in a plot.
p = figure(title="Iris Dataset", x_axis_label="Petal Length", y_axis_label="Petal Width")
p.circle(flowers["petal_length"], flowers["petal_width"], size=10, color="blue")

# Set the output file name
output_file("scatter.html")

# Display the plot
show(p)
```

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
```

```python
x = [3,4,5,6,7]
y = [4,5,6,7,4]
output_file("line.html")
p = figure(title = "line graph")
p.scatter(x,y , size = 20,fill_color = 'red', legend_label = 'this is my data point')
show(p)
```

```python
# Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time
```

```python
#The Bokeh server is a key feature of the Bokeh library that allows you to create and serve interactive plots that can be updated in real time. 
#With the Bokeh server, you can create dynamic web applications that allow users to interact with and explore your data in real time.

#To use the Bokeh server, you first need to create a Python script that defines your interactive plot. This script should include a curdoc() function call, which creates a 
#Bokeh document object that represents your plot.Once you have defined your plot, you can start the Bokeh server by running the bokeh serve command in a terminal window. 
#This will launch a Bokeh server that listens for incoming connections on a specified port.

#To view your interactive plot, you can open a web browser and navigate to the URL of the Bokeh server. This will display your plot in the browser window, and users can interact 
#with the plot by clicking or dragging the mouse.
#as users interact with the plot, the Bokeh server updates the plot in real time, using a websocket connection to send updates to the browser. 
#This allows users to explore the data and see changes to the plot in real time, without having to reload the page.
```
